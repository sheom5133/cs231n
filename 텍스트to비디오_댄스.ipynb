```python
from PIL import Image
import numpy as np
import torch

def to_pil_frames(video_frames):
    """
    video_frames가 다음 중 하나일 때:
      - 리스트: 각 요소가 PIL.Image, 또는 numpy.ndarray (H,W), (H,W,1), (H,W,3), (H,W,4), 또는 torch.Tensor
      - numpy.ndarray: 4D 배열 (N,H,W,C) 또는 3D 배열 (H,W,C) 혹은 (H,W)
      - torch.Tensor: 4D (N,C,H,W) 또는 3D (C,H,W) 또는 2D (H,W)
    모두 PIL RGB 이미지 리스트로 변환하여 반환.
    """
    pil_list = []

    # 1) Torch Tensor인 경우 NumPy 배열로 변환
    if isinstance(video_frames, torch.Tensor):
        # Tensor shape: (N,C,H,W) 또는 (C,H,W) 또는 (H,W)
        arr = video_frames
        # CPU로 이동 및 numpy 변환
        arr = arr.detach().cpu().numpy()
        video_frames = arr  # 다음 분기에서 처리하게 덮어쓰기

    # 2) NumPy ndarray인 경우
    if isinstance(video_frames, np.ndarray):
        arr = video_frames
        # 4D array: (N,H,W,C) 또는 (N,C,H,W)? 모델마다 다를 수 있음. 우선 (N,H,W,C) 형태 가정
        if arr.ndim == 4:
            # 두 가능성: (N,H,W,C) 또는 (N,C,H,W)
            N, a, b, c = arr.shape
            # 간단 판단: 채널 수 위치 확인
            if c in (1,3,4):
                seq = [arr[i] for i in range(N)]
            elif a in (1,3,4):
                # (N,C,H,W) 형태인 경우: 바꿔 담기
                seq = [np.moveaxis(arr[i], 0, -1) for i in range(N)]
            else:
                raise ValueError(f"Unsupported 4D numpy array shape for frames: {arr.shape}")
        elif arr.ndim == 3:
            # (H,W,C) 혹은 (C,H,W)
            H, W, C = arr.shape
            if C in (1,3,4):
                seq = [arr]
            elif H in (1,3,4):
                # (C,H,W) 형태일 수 있음: 채널 축 앞으로 옮겨서 (H,W,C)
                seq = [np.moveaxis(arr, 0, -1)]
            else:
                raise ValueError(f"Unsupported 3D numpy array shape for single frame: {arr.shape}")
        elif arr.ndim == 2:
            # (H,W) 흑백 한 프레임
            seq = [arr]
        else:
            raise ValueError(f"Unsupported numpy array ndim for frames: {arr.shape}")
    elif isinstance(video_frames, list):
        # 이미 리스트라면 내부 요소들을 seq로 사용
        seq = video_frames
    else:
        raise ValueError(f"Unsupported video_frames type: {type(video_frames)}")

    # 3) 각 프레임 요소를 PIL RGB 이미지로 변환
    for idx, f in enumerate(seq):
        # 3-1) PIL 이미지인 경우
        if isinstance(f, Image.Image):
            img = f.convert("RGB")
            pil_list.append(img)
            continue

        # 3-2) Torch Tensor인 경우: 이미 NumPy 변환되었으면 여기 해당 안 됨

        # 3-3) NumPy 배열인 경우
        if isinstance(f, np.ndarray):
            arr = f
            # dtype이 uint8이 아닐 수 있으므로 안전하게 변환
            if arr.dtype != np.uint8:
                # float인 경우 0~1 범위라면 255 곱하고 uint8
                if np.issubdtype(arr.dtype, np.floating):
                    arr = np.clip(arr, 0.0, 1.0)
                    arr = (arr * 255).astype(np.uint8)
                else:
                    # 정수 타입 등은 바로 uint8 캐스팅
                    arr = arr.astype(np.uint8)
            # 배열 shape에 따라 채널 변환
            if arr.ndim == 2:
                # (H,W) -> (H,W,3)
                arr = np.stack([arr]*3, axis=-1)
            elif arr.ndim == 3:
                H, W, C = arr.shape
                if C == 1:
                    arr = np.concatenate([arr]*3, axis=2)
                elif C == 3:
                    pass
                elif C == 4:
                    # RGBA 형태라면 알파 제거
                    arr = arr[:, :, :3]
                else:
                    raise ValueError(f"Unsupported numpy frame channels at index {idx}: shape={arr.shape}")
            else:
                raise ValueError(f"Unsupported numpy frame ndim at index {idx}: shape={arr.shape}")
            # PIL 이미지로 변환
            img = Image.fromarray(arr)
            img = img.convert("RGB")
            pil_list.append(img)
            continue

        # 3-4) 그 외 타입: 예를 들어 torch.Tensor가 리스트 내부에 남아있다면 처리
        if isinstance(f, torch.Tensor):
            arr = f.detach().cpu().numpy()
            # 위와 동일 변환 로직 재사용
            if arr.dtype != np.uint8:
                if np.issubdtype(arr.dtype, np.floating):
                    arr = np.clip(arr, 0.0, 1.0)
                    arr = (arr * 255).astype(np.uint8)
                else:
                    arr = arr.astype(np.uint8)
            if arr.ndim == 2:
                arr = np.stack([arr]*3, axis=-1)
            elif arr.ndim == 3:
                H, W, C = arr.shape
                if C == 1:
                    arr = np.concatenate([arr]*3, axis=2)
                elif C == 3:
                    pass
                elif C == 4:
                    arr = arr[:, :, :3]
                else:
                    raise ValueError(f"Unsupported tensor frame channels at index {idx}: shape={arr.shape}")
            else:
                raise ValueError(f"Unsupported tensor frame ndim at index {idx}: shape={arr.shape}")
            img = Image.fromarray(arr)
            img = img.convert("RGB")
            pil_list.append(img)
            continue

        # 3-5) 지원되지 않는 타입
        raise ValueError(f"Unsupported frame element type at index {idx}: {type(f)}")

    return pil_list

# 사용 예시:
# video_frames 변수가 이미 존재한다고 가정
pil_frames = to_pil_frames(video_frames)
# 변환되었는지 확인
for i, img in enumerate(pil_frames[:5]):  # 앞 5프레임만 확인
    print(f"PIL Frame {i}: mode={img.mode}, size={img.size}")
# 이후 export_to_video 등에 넘김
from diffusers.utils import export_to_video
from IPython.display import HTML, display

video_path = export_to_video(pil_frames)
print("Video saved at:", video_path)
display(HTML(f'<video width=480 controls src="{video_path}"></video>'))
```

## video_frames가 다음 중 하나일 때:
###       - 리스트: 각 요소가 PIL.Image, 또는 numpy.ndarray, 또는 torch.Tensor
###       - numpy.ndarray: 5D (B, N, H, W, C) 또는 4D (N, H, W, C)/(B, C, H, W) 등
###                       또는 3D/2D(single frame)
###       - torch.Tensor: 5D (B, N, C, H, W) 또는 4D (N, C, H, W)/(C, H, W)/(H, W)
###     모두 PIL RGB 이미지 리스트로 변환하여 반환.
  

```python
from PIL import Image
import numpy as np
import torch

def to_pil_frames(video_frames):
    """
    video_frames가 다음 중 하나일 때:
      - 리스트: 각 요소가 PIL.Image, 또는 numpy.ndarray, 또는 torch.Tensor
      - numpy.ndarray: 5D (B, N, H, W, C) 또는 4D (N, H, W, C)/(B, C, H, W) 등
                      또는 3D/2D(single frame)
      - torch.Tensor: 5D (B, N, C, H, W) 또는 4D (N, C, H, W)/(C, H, W)/(H, W)
    모두 PIL RGB 이미지 리스트로 변환하여 반환.
    """
    pil_list = []

    # 1) Torch Tensor인 경우 numpy로 변환
    if isinstance(video_frames, torch.Tensor):
        arr = video_frames.detach().cpu().numpy()
        video_frames = arr

    # 2) NumPy ndarray인 경우
    if isinstance(video_frames, np.ndarray):
        arr = video_frames
        # 5D 배열: (B, N, H, W, C) 또는 (B, N, C, H, W)
        if arr.ndim == 5:
            B, d1, d2, d3, d4 = arr.shape
            # 경우 1: (B, N, H, W, C) where C in {1,3,4}
            if d4 in (1, 3, 4):
                seq = []
                for b in range(B):
                    for i in range(d1):
                        seq.append(arr[b, i])  # shape (H, W, C)
            # 경우 2: (B, N, C, H, W) where C in {1,3,4}
            elif d1 in (1, 3, 4):
                # Interpret shape as (B, C, H, W, ?)? Unlikely.
                # More common is (B, N, C, H, W) so d2,d3 are H,W
                # But our unpack: B,d1,d2,d3,d4. If d1 in {1,3,4} treat as C, so shape (B, C, H, W, N)? Hard to guess.
                # Better check shapes: we assume (B, N, C, H, W) if arr.shape == (B, N, C, H, W)
                # In that case: d1=N? But N likely >4. So fallback:
                # If arr shape (B, N, C, H, W), then d2 (d2=H) and d3 (d3=W) and d4 invalid.
                # Let's detect if arr shape matches (B, N, C, H, W): check arr.shape[2] in {1,3,4}
                # Actually arr.shape: (B, N, C, H, W): arr.ndim==5, arr.shape[2] in {1,3,4}.
                # But our unpack did (B,d1,d2,d3,d4) = arr.shape, so we want to detect arr.shape[2] in {1,3,4}:
                if arr.shape[2] in (1,3,4):
                    seq = []
                    for b in range(B):
                        for i in range(arr.shape[1]):  # N
                            frame_arr = arr[b, i]  # shape (C, H, W)
                            # move channel last:
                            frame_arr = np.moveaxis(frame_arr, 0, -1)  # (H, W, C)
                            seq.append(frame_arr)
                else:
                    raise ValueError(f"Unsupported 5D numpy array shape for frames: {arr.shape}")
            else:
                raise ValueError(f"Unsupported 5D numpy array shape for frames: {arr.shape}")

        # 4D 배열: (N, H, W, C) 또는 (B, C, H, W) 또는 (N, C, H, W)
        elif arr.ndim == 4:
            # 경우 A: (N, H, W, C) where C in {1,3,4}
            N, a, b, c = arr.shape
            if c in (1,3,4):
                seq = [arr[i] for i in range(N)]
            # 경우 B: (B, C, H, W) or (N, C, H, W)
            elif a in (1,3,4):
                # Interpret as (N, C, H, W) or (B, C, H, W). Flatten batch if B>1?
                # 여기서는 (N, C, H, W)로 간주
                seq = []
                for i in range(arr.shape[0]):
                    frame_arr = arr[i]  # shape (C, H, W)
                    frame_arr = np.moveaxis(frame_arr, 0, -1)  # (H, W, C)
                    seq.append(frame_arr)
            else:
                raise ValueError(f"Unsupported 4D numpy array shape for frames: {arr.shape}")

        # 3D 배열: (H, W, C) 또는 (C, H, W)
        elif arr.ndim == 3:
            H, W, C = arr.shape
            if C in (1,3,4):
                seq = [arr]
            elif H in (1,3,4):
                # (C, H, W)로 간주
                frame_arr = np.moveaxis(arr, 0, -1)
                seq = [frame_arr]
            else:
                raise ValueError(f"Unsupported 3D numpy array shape for single frame: {arr.shape}")

        # 2D 배열: (H, W)
        elif arr.ndim == 2:
            seq = [arr]
        else:
            raise ValueError(f"Unsupported numpy array ndim for frames: {arr.shape}")

    elif isinstance(video_frames, list):
        seq = video_frames
    else:
        raise ValueError(f"Unsupported video_frames type: {type(video_frames)}")

    # 3) 각 요소를 PIL RGB 이미지로 변환
    for idx, f in enumerate(seq):
        # PIL.Image인 경우
        if isinstance(f, Image.Image):
            img = f.convert("RGB")
            pil_list.append(img)
            continue
        # Torch Tensor인 경우 (남아 있다면)
        if isinstance(f, torch.Tensor):
            arr = f.detach().cpu().numpy()
        else:
            arr = f

        # NumPy 배열인 경우
        if isinstance(arr, np.ndarray):
            # dtype 처리
            if arr.dtype != np.uint8:
                if np.issubdtype(arr.dtype, np.floating):
                    arr = np.clip(arr, 0.0, 1.0)
                    arr = (arr * 255).astype(np.uint8)
                else:
                    arr = arr.astype(np.uint8)
            # 차원 처리
            if arr.ndim == 2:
                arr = np.stack([arr]*3, axis=-1)
            elif arr.ndim == 3:
                H, W, C = arr.shape
                if C == 1:
                    arr = np.concatenate([arr]*3, axis=2)
                elif C == 3:
                    pass
                elif C == 4:
                    arr = arr[:, :, :3]
                else:
                    raise ValueError(f"Unsupported numpy frame channels at index {idx}: shape={arr.shape}")
            else:
                raise ValueError(f"Unsupported numpy frame ndim at index {idx}: shape={arr.shape}")
            img = Image.fromarray(arr)
            img = img.convert("RGB")
            pil_list.append(img)
            continue

        # 지원되지 않는 타입
        raise ValueError(f"Unsupported frame element type at index {idx}: {type(f)}")

    return pil_list


# 사용 예시:
# video_frames 변수가 이미 존재한다고 가정
# 예를 들어 출력된 shape이 (1,16,256,256,3)인 경우:
# video_frames = np.random.randint(0,255,(1,16,256,256,3),dtype=np.uint8)

try:
    pil_frames = to_pil_frames(video_frames)
    for i, img in enumerate(pil_frames[:5]):
        print(f"PIL Frame {i}: mode={img.mode}, size={img.size}")
    from diffusers.utils import export_to_video
    from IPython.display import HTML, display
    video_path = export_to_video(pil_frames)
    print("Video saved at:", video_path)
    display(HTML(f'<video width=480 controls src="{video_path}"></video>'))
except Exception as e:
    print("변환 중 오류 발생:", e)
```
